# Importing all the necessary libraries 

```python
import pandas as pd
import numpy as np
import plotly.graph_objects as go
```

# Importing all the data.

```python
aws_county = pd.read_csv(r'C:\Users\Akshay Deshmukh\Desktop\MS Studies\CS682 Project\3D Visualization\Datasets\AWS county level.csv')

counties = ['Barnstable','Berkshire','Bristol','Dukes','Essex','Franklin','Hampden','Hampshire','Middlesex','Nantucket','Norfolk','Plymouth','Suffolk','Worcester']
```

# Function to plot county wise surface plots.

```python
def surface_plot(data, county):
    
    data = data[data.name.isin(counties)]                       # Selecting data for all the counties in Massachusetts.
    
    
    data = data[['name', 'avg_risk_score_all']]                 # Keeping the necessary columns.
    
    data['name'] = data['name'].apply(lambda X: X.lower())
    
    if county not in data['name'].unique() :     
        return  print("Data for county not available.")    # Condition to check if the input city is a valid city in the US.
    
    data = data[data['name']== county.lower()]                  # Selecting/Filtering the data for the input county.
    
    risk_list = list(data.avg_risk_score_all)                   # Making the list of avg risks for the input county.
    
    SLR = []
    for i in risk_list:
        SLR.append(np.linspace(0, i, 100))                      # Create evenly spaced numbers for the risk.            
    
    df = pd.DataFrame(SLR)                                      
    
    fig = go.Figure(data=[go.Surface(z=df.values)])             # Plotting the surface plot.
    
    fig.update_layout(title = 'Average risk for ' + county.upper() + ' COUNTY.',
                      scene = dict(
                        zaxis_title='Avg risk.',
                        xaxis = dict(
                            ticktext = [''],
                              tickvals = [0]), 
                        yaxis = dict(
                            ticktext = [''],
                              tickvals = [0])),
                        width=900,
                        margin=dict(l=10, r=10, b=40, t=40)    # Editing the layout of the 3D plot.
                      )
    
    return fig.show()    
```

# Implementation.

```python
surface_plot(aws_county, 'suffolk')
```
